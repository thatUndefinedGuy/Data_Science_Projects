```python
import pandas as pd
import seaborn as sns 
import random
from sklearn import linear_model as lm
import matplotlib as mpl
```

```python
m = 3 #float(input("slope")) 
b = 8 #int(input("inter")) 
```

```python
x = []
y = []  
for i in range(100): 
    x.append(i/10+ random.uniform(-1,1)) 
    y.append(((i/10) * m) + (b + random.uniform(-15,15))) 
df = pd.DataFrame(data={"X":x,"Y":y}) 
sns.regplot(df,x="X",y="Y")
ln = lm.LinearRegression()
ln.fit(df[["X"]],df[["Y"]])
mess = "y="+str(ln.coef_)+"x + "+str(ln.intercept_)
mess
print(ln.score(df[["X"]],df[["Y"]]))
```

```python
pre = ln.predict(df[["X"]])
xvals = []
yvals = []
for i in pre[0:50]:
    xvals.append(int(i))
    yvals.append(int(i) * m+b)
df2 = pd.DataFrame({"x":xvals,"y":yvals})
sns.scatterplot(df,x="X",y="Y")
sns.scatterplot(df2,x = "x",y="y")
df2.plot(kind="line")
```
